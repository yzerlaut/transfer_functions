```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg, special
```

```python
mu, theta, sigma = 1, .3,1
dt = 0.1
```

```python
def cov_matrix(sigma, theta, tmax, dt=dt):
    """calculate and return covariance matrix of ornstein-uhlenbeck proccess"""
    time_lag = np.arange(int(tmax/dt))*dt
    column = sigma**2/(2.*theta)*np.exp(-theta*time_lag)
    cov = linalg.toeplitz(column)
    return cov

def rotate_matrix(vec, mat):
    return np.dot(np.dot(vec, mat), vec.T)

def gaussian_1d(x, mu, sigma):
    return 1./(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2.*sigma**2))

def prob_threshold_crossing(theta, mu, sigma):
    return special.erfc((theta-mu)/(np.sqrt(2)*sigma))/2.

def calculate_cov_matrix(y, tmax, dt=dt):
    y_nomean = y - np.mean(y)
    n_max = int(tmax/dt)
    n_pts = len(y_nomean)
    cov = [np.mean(y_nomean[i:]*y_nomean[:n_pts-i])
          for i in range(n_max)]
    return linalg.toeplitz(cov)
    
```

```python
def ornstein_uhlenbeck(mu, sigma, theta, tmax, dt=dt):
    """return a realisation of ornstein-uhlenbeck process"""

    diffcoef = sigma**2
    y0 = mu
    tau = 1./theta
    
    n_steps = int(tmax/dt)
    
    A = np.sqrt(diffcoef*tau/2.*(1-np.exp(-2*dt/tau)))

    noise = np.random.randn(n_steps)

    y = np.zeros(n_steps)
    y[0] = y0

    for i in range(n_steps-1):
        y[i+1] = y0 + (y[i]-y0)*np.exp(-dt/tau)+A*noise[i]
    return y
```

```python
def compare_theory_simulation(mu, theta, sigma, threshold):
    npts = 50000
    ou_samples = ornstein_uhlenbeck(mu, sigma, theta, npts)
    spikes = np.convolve(ou_samples, G_filter)>threshold
    spike_rate = np.sum(spikes)*1./len(spikes)
    
    cov_theory = cov_matrix(sigma, theta, filter_length)
    sigma_proj = np.sqrt(rotate_matrix(G_filter[::-1], cov_theory))
    mu_proj = np.sum(G_filter)*mu
    prob_theory = prob_threshold_crossing(threshold, mu_proj, sigma_proj)
    
    return spike_rate, prob_theory

def plot_comparison(value_list, title=None, labels=True):
    simulation, theory = np.array(value_list).T
    plt.plot(simulation, theory, 'o')
    xmin, xmax = plt.xlim()
    ymin, ymax = plt.ylim()
    mmax = max(xmax, ymax)
    mmin = min(xmin, ymin)
    plt.plot([mmin, mmax], [mmin, mmax], 'k-')
    plt.xticks([mmin, mmax])
    plt.yticks([mmin, mmax])
    if labels:
        plt.xlabel('spike rate - simulation',labelpad=-8)
        plt.ylabel('spike rate - theory', labelpad=-8)
        
    ax = plt.gca()
    if title:
        plt.text(0.1, 0.8, title, transform=ax.transAxes)
    plt.xlim([mmin, mmax])
    plt.ylim([mmin, mmax])
```

Compute the spike-triggered average and whiten it using the whitening transformation $z=Vx$, where $V=E D^{-\frac{1}{2}} E^T$. ($E$ is the matrix of eigenvectors aranged in columns and $D$ is the diagonal matrix of eigenvalues of covariance matrix of $x$)

```python
mu, sigma, tau = 1., 15., 5.
dt = 0.5
threshold = 3.
def rescaled_ou(mu, sigma,  tau, tmax, dt=dt): 
    return ornstein_uhlenbeck(mu, sigma*np.sqrt(2./tau), 1./tau, tmax, dt)

def spike_train(samples, threshold, dt):
    return (samples[1:]>threshold) & (samples[:-1]<threshold)
    #return (samples[1:]-samples[:-1])>threshold*dt

def calc_fr(mu, sigma, tau, threshold, tmax=1000, dt=dt):
    ou_samples = rescaled_ou(mu, sigma, tau, tmax, dt)
    rate = np.mean(spike_train(ou_samples, threshold, dt))/np.sqrt(dt)
    return rate
    
#plt.plot(rescaled_ou(mu, sigma, tau, 100))
print calc_fr(mu, sigma, tau, threshold, 100, dt)
print calc_fr(mu, sigma, tau, threshold, 100, dt/10.)
```

```python
from scipy import integrate

def gaussian_nd(x, mu, cxx):
    d=len(x)
    norm = np.sqrt((2*np.pi)**(d)*np.linalg.det(cxx))
    
    inv_cxx = np.linalg.inv(cxx)
    delta_x  = x - mu
    exponent = -1/2.*np.dot(np.dot(delta_x, inv_cxx), delta_x)
    return 1./norm*np.exp(exponent)

def ou_rate(mu, sigma, tau, threshold, dt=dt):
    cxx = cov_matrix(sigma*np.sqrt(2./tau), 1./tau, 2*dt, dt)
    prob = lambda x, y: gaussian_nd(np.array([x,y]), mu, cxx)

    gfun = lambda x: threshold
    hfun = lambda x: np.inf
    p, err = integrate.dblquad(prob, -np.inf, threshold, gfun, hfun)
    
    return p/np.sqrt(dt)

rate_theory = ou_rate(mu, sigma, tau, threshold)
rate_simulations = calc_fr(mu, sigma, tau, threshold)
print "Firing rate theory: %.2f, simulations: %.2f" % (rate_theory, rate_simulations)
```

```python
from functools import partial

def fr_tau(mu, sigma, taus, threshold):
    def _wrapper(f):
        return np.array([f(mu, sigma, tau, threshold) for tau in taus])
    return _wrapper

taus = np.linspace(0.1, 4, 20)
funcs = [partial(calc_fr, tmax=1e3), ou_rate]

fr_vs_tau, fr_theory = map(fr_tau(mu, sigma, taus, threshold), funcs)
plt.plot(taus, fr_vs_tau, 'r.', label='threshold=%.2f' % threshold)
plt.plot(taus, fr_theory, 'r-', label='theory' )

fr_vs_tau, fr_theory = map(fr_tau(mu, sigma, taus, threshold*0.5), funcs)
plt.plot(taus, fr_vs_tau, 'b.', label='threshold=%.2f' % (threshold*0.5))
plt.plot(taus, fr_theory, 'b-', label='theory' )

#tau_arr, fr_vs_tau = np.array([(_, calc_fr(mu, sigma, _, threshold*0.9, n=1e2)) for _ in np.linspace(0.1, 4, 20) ]).T
#plt.plot(tau_arr, fr_vs_tau, '.', label='threshold=%.2f' % (threshold*0.9))
#plt.plot(tau_arr, fr_theory, 'r-', label='theory' )

plt.legend()
```

```python
from scipy import special

def gauss_integral(x, mu, sigma):
    """integral of normal distribution from x to inf"""
    return special.erfc((x-mu)/np.sqrt(2)/sigma)/2.

def ou_rate_rotated(mu, sigma, tau, threshold, dt=0.1):
    """Calculate OU rate in rotated space"""
    
    #calculate 2x2 auto-covariance matrix of the original process
    cxx = cov_matrix(sigma*np.sqrt(2./tau), 1./tau, 2*dt, dt)
    
    #find (unitary) projections orthogonalising the covariance matrix
    _, trans_matrix = np.linalg.eig(cxx)
    
    #calculate the crossing integration limits
    x_pos = np.dot(trans_matrix, np.array([threshold, threshold]))
    
    #reorder projection vectors such that x_pos[1] == 0
    vec_order = x_pos.argsort()[::-1]
    trans_matrix = trans_matrix[vec_order, :]
    x_pos = np.dot(trans_matrix, np.array([threshold, threshold]))
    
    #calculate std. deviations and means in the rotated space
    rot_cxx = rotate_matrix(trans_matrix, cxx)
    sigma2_eff = np.diag(rot_cxx)
    mu_eff = np.dot(trans_matrix, np.ones(2)*mu)
    
    #define the limits of integration from "threshold limits" in original space
    A = trans_matrix.T
    ulimit = lambda x: A[0,1]/A[0,0]*x + x_pos[0]
    llimit = lambda x: A[1,1]/A[1,0]*x + x_pos[0]
    
    #integrate in the rotated space
    #  probability distribution factorises after rotation!
    integrand = lambda x: ((gauss_integral(llimit(x), mu_eff[0], np.sqrt(sigma2_eff[0]))-
                             gauss_integral(ulimit(x), mu_eff[0], np.sqrt(sigma2_eff[0])))*
                             gaussian_1d(x, mu_eff[1], np.sqrt(sigma2_eff[1])))
    
    return (integrate.quad(integrand, x_pos[1], np.inf)[0])/np.sqrt(dt)
```
