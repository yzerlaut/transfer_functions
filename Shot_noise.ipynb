# constructing the shotnoise experiment
see https://github.com/yzerlaut/tf_review/blob/master/BrunelSergi98.ipynb for details and further analysis

```python
def PSP(t, Tm, Ts, I0, V0=0):
    if Ts==0:
       return (V0+I0)*np.exp(-t/Tm) 
    if Tm!=Ts:
        return (V0+I0*Tm/(Tm-Ts))*np.exp(-t/Tm)-I0*Tm/(Tm-Ts)*np.exp(-t/Ts)
    else:
        return (V0+I0*t/Tm) * np.exp(-t/Tm)
def Tmax(Tm, Ts, I0, V0=0.): # delay of the max jump value after a synaptic event
    if Ts==0:
        Tmax=0
    elif Tm!=Ts: 
        Tmax = Tm*Ts/(Tm-Ts) * np.log( Tm**2*I0 / ( Ts*(Tm-Ts)*(V0+Tm*I0/(Tm-Ts)) ) )
    else:
        Tmax = Tm*(1-V0/I0)
    return Tmax

from scipy.optimize import newton

def find_threshold_crossing(Vthre, Tm, Ts, I0, V0=0.):
    # to be executed only if there is an actual threshold crossing V(tmax)>Vth
    if Ts==0:
        res=0 # in this case, the spike is instantaneous
    elif Tm!=Ts:
        fun = lambda x:  (V0+I0*Tm/(Tm-Ts))*np.exp(-x/Tm)-I0*Tm/(Tm-Ts)*np.exp(-x/Ts) -Vthre
        res = newton(fun, 0.)#, method='SLSQP', bounds=bnds)
    else:
        fun = lambda x:  (V0+I0*x/Tm) * np.exp(-x/Tm) -Vthre
        res = newton(fun, 0.)#, method='SLSQP', bounds=bnds)
    return res

def Shotnoise_Sim(nu, J, Tm, Ts, v_thre, v_reset, max_event=1e5):
    if nu==0: print "NON ZERO FREQUENCY NEEDED!!"
    
    # we prepare the pre-synaptic spike train
    pre_spk_train = np.random.exponential(1./nu, max_event)
    
    counter_spk = 0 # spike counter
    i, v = 0., v_reset # initial conditions
    t_last = pre_spk_train[0] # time at last pre-syn event
    for pre_spk in pre_spk_train[1:]:
        # we start by testing the effect of the last spike
        i+=J # there has been an increment of the input
        # did this last event could trigger a spike ??
        if PSP(Tmax(Tm, Ts, i, V0=v), Tm, Ts, i, V0=v)>v_thre:
            # if it can, when  would it trigger it ?
            tspk = find_threshold_crossing(v_thre, Tm, Ts, i, V0=v)
        else:    
            tspk = np.inf # this means no post-syn spike !
        # so now, we test if there is no pre-syn spike, that happens before the post-syn spike
        if tspk<pre_spk: # then SPIKE
            counter_spk +=1
            v = v_reset # we reset the potential
            #print tspk
        else:
            v = PSP(pre_spk, Tm, Ts, i, V0=v)  # we update the potential
        # then, whatever happens, we update the time to the next spike, as nothing can happen without any event
        # (no leak because El = v_reset)
        if Ts!=0:
            i = i*np.exp(-pre_spk/Ts)
        else:
            i=0
        t_last += pre_spk
    return counter_spk/np.cumsum(pre_spk_train)[-1]
```

```python
## running the simulation

Tm = 1. # 20ms
Ts = 0.2 
DV = 0.01 # 0.1 mV
vthre = 1
vrest = 0
nu_sn = np.arange(80,101,1)
fr_sn = np.array([Shotnoise_Sim(_, DV, Tm, Ts, vthre, vrest, max_event=1e5) for _ in nu_sn]).T
plt.plot(nu_sn,fr_sn,'D')
```

# Stochastic Process stimulation

```python
%%cython
cimport cython
from libc.math cimport exp, sqrt, pow, log, erf, cos
from libc.stdlib cimport rand, RAND_MAX

cdef double PI = 3.14159265358979323846

@cython.cdivision(True)
cdef double drand():
    return (rand()+1.0)/(RAND_MAX+1.0);

@cython.cdivision(True)
cdef double random_normal():
    return sqrt(-2*log(drand())) * cos(2*PI*drand());


@cython.cdivision(True)
cdef double ou_sample(double y, double dt, double mu, double A, double tau):

    cdef double noise = random_normal()
    y = mu + (y - mu) * exp(-dt/tau) + A * noise
    return y

@cython.cdivision(True)
def fast_integrate_reset(double tmax, double dt, double v_thre, double v_reset, double tau, 
                    double imu, double isigma, double itau):
    cdef int max_steps = (int)(tmax/dt)
    cdef double diffcoef = 2*isigma**2/itau
    cdef double A = sqrt(diffcoef*itau/2.*(1-exp(-2*dt/itau)))
    
    spikes, mem_potential = [], []
    cdef double v = v_reset

    cdef double t
    cdef int i
    cdef double current = imu
    for i in range(max_steps):
        t = i*dt
        current = ou_sample(current, dt, imu, A, itau)
        v_old = v
        v = v - v*dt/tau + current*dt
        #v = current
        #v = ou_sample(v, dt, imu, A, itau)
        if (v>v_thre) and (v_old<v_thre):
            spikes.append(t)
            #v = v_reset
        mem_potential.append(v)
    return spikes, mem_potential
```

```python
def fast_calc_fr(tmax, dt, mu, sigma, tau, tau_m, threshold):
    spikes, _ = fast_integrate_reset(tmax, dt, threshold, 0, tau_m, mu, sigma, tau) 
    t_trans = 1.
    spikes = np.array(spikes)
    #integrate_reset(tmax, dt, gexc, threshold, mu, store=False)
    return np.sum(spikes>t_trans)/(tmax-t_trans)
```

```python
## running the simulation

Tm = 1. # 20ms
Ts = 0.2 # 
DV = 0.01 # 0.1 mV
vthre = 1
vrest = 0
dt = 1e-3

nu_ou = np.linspace(80,100)
tmax = 1e3
param, fr_ou = np.array([(_, fast_calc_fr(tmax, dt, _*Tm*DV, DV*np.sqrt(Tm*_), Ts, Tm, vthre)) for _ in nu_ou]).T
plt.plot(nu_ou, fr_ou,'D')
```

# Analytical estimation

```python
# btel estimates
from scipy import special, integrate

def gauss_integral(x, mu, sigma):
    """integral of normal distribution from x to inf"""
    return special.erfc((x-mu)/np.sqrt(2)/sigma)/2.

def gaussian_1d(x, mu, sigma):
    return 1./(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2.*sigma**2))

def erf_integral(x, mu, sigma):
    """integral from -inf to x of gauss_integral"""
    x_ = (x-mu)/sigma/np.sqrt(2)
    return (x_*(1+special.erf(x_))+np.exp(-x_**2)/np.sqrt(np.pi))*sigma*np.sqrt(2)

def ou_rate_markov_anal(mu, sigma, tau_m, tau, threshold):
    sigma_markov = np.sqrt(sigma**2/(1./tau_m*(1./tau_m+1./tau)))
    mu_markov = (mu*tau_m)
    correction = np.sqrt((tau+tau_m)/tau_m)
    
    stationary = gaussian_1d(threshold, mu_markov, sigma_markov)
    markov_integral = erf_integral(0, (threshold), sigma)/2.
    prob = markov_integral*stationary/correction
    return prob

def ou_rate_markov_anal2(mu, sigma, tau_m, tau, threshold):
    sigma_markov = np.sqrt(sigma**2/(1./tau_m*(1./tau_m+1./tau)))
    mu_markov = (mu*tau_m)
    
    stationary = gaussian_1d(threshold, mu_markov, sigma_markov)
    markov_integral = erf_integral(0, (threshold-mu_markov)/np.sqrt(tau), sigma_markov/tau)/2.
    prob = markov_integral*stationary
    return prob
```

```python
# brunel estimates

import scipy.integrate as sp_int
import scipy.special as sp_spec
def Psi(Array):
    return np.exp(Array*Array)*(sp_spec.erf(Array)+1)

## First passage time see. e.g. Brunel & Hakim, 1998
def Brunel_function(nu, J, tau, El, Vthre, Vreset):
    x1=(Vreset-El-nu*J*tau)/(J*np.sqrt(nu*tau))
    x2=(Vthre-El-nu*J*tau)/(J*np.sqrt(nu*tau))
    integral = sp_int.quad(Psi,x1,x2)
    if integral[0]<float('inf') and integral[0]>0:
        T=Tm*np.sqrt(np.pi)*integral[0]
        freq=1.0/T
    else:
        freq=1e-9 # arbitrary for "negligible"
    return freq

## now with correction for correlated noise input 
# (membrane equation with colored noise input, correlation time : Tau )
def BrunelFunc_colored(nu, J, tau, El, Vthre, Vreset, Ts):
    # taken from Brunel and Sergi 1998 :
    mu = J*nu*tau
    s = J*np.sqrt(nu*tau)
    theta_hat = np.sqrt(2)*(Vthre-mu)/s
    theta = Vthre + s*( 1.03*np.sqrt(Ts/Tm) - Ts/Tm*(-0.34 + 0.29*theta_hat) )
    return Brunel_function(nu, J, tau, El, theta, Vreset)
```

```python
# kuhn's estimate

def KuhnFunc(mVm, sVm, Tm, Vthre):
    return sp_spec.erfc( (Vthre - mVm)/sVm/np.sqrt(2) )/2/Tm
```

```python
## Numerical Evaluation

nu_th = np.linspace(80,100,1e2,endpoint=True)

btel0 = ou_rate_markov_anal(nu_th*Tm*DV, DV*np.sqrt(nu_th*Tm), Tm, Ts, vthre)
brunel0 = np.array([(Brunel_function(_, DV, Tm, vrest, vthre, vrest)) for _ in nu_th]).T    
brunel1 = np.array([(BrunelFunc_colored(_, DV, Tm, vrest, vthre, vrest, Ts)) for _ in nu_th]).T   
kuhn0 = KuhnFunc(nu_th*Tm*DV, DV*np.sqrt(nu_th*Tm), Tm, vthre)
```

# Running the whole thing to get this figure !

```python
def WholeThing(DV, Tm, Ts, vthre=1, vrest=0, dt = 1e-3, fmin=70, fmax=100):
    
    # exp frequencies
    nu_exp = np.linspace(fmin, fmax, 20, endpoint=True)
    # shotnoise
    fr_sn = np.array([Shotnoise_Sim(_, DV, Tm, Ts, vthre, vrest, max_event=1e5) for _ in nu_exp]).T
    # stochstic process
    tmax = 1e6*dt   
    fr_ou = np.array([fast_calc_fr(tmax, dt, _*Tm*DV, DV*np.sqrt(Tm*_), Ts, Tm, vthre) for _ in nu_exp]).T

    # theoretical
    nu_th = np.linspace(fmin,fmax,1e2,endpoint=True)
    btel0 = ou_rate_markov_anal(nu_th*Tm*DV, DV*np.sqrt(nu_th*Tm), Tm, Ts, vthre)
    brunel1 = np.array([(BrunelFunc_colored(_, DV, Tm, vrest, vthre, vrest, Ts)) for _ in nu_th]).T   
    kuhn0 = KuhnFunc(nu_th*Tm*DV, DV*np.sqrt(nu_th*Tm), Tm, vthre)

    # figure
    plt.figure(figsize=(10,6)) 
    plt.title(r" $\tau_S / \tau_m $ = "+str(round(Ts/Tm,1))+ \
    ",    $ \quad \Delta_V / (V_{thre}-V_{reset})$ = "+str(round(100*DV/(vthre-vrest),1))+" $\%$" )
    plt.plot(nu_exp, fr_sn,'D', label=' Shot Noise')
    plt.plot(nu_exp, fr_ou,'D', label=' Stochastic process')
         
    plt.plot(nu_th, btel0, label = 'btel approx', lw=3)
    plt.plot(nu_th, brunel1, label = 'Brunel \& Sergi', lw=3)
    plt.plot(nu_th, kuhn0, label = 'Kuhn et al.', lw=3)
    
    plt.legend(loc='upper left')
    plt.xlabel(r"$\nu_{in}\,.\,\tau$", fontsize=26)
    plt.ylabel(r"$\nu_{out}\,.\,\tau$", fontsize=26)
```

```python
DV = 0.01
Tm = 1.
Ts = 0.2
WholeThing(DV, Tm, Ts, fmin=80)
```
