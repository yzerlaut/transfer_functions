```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import transforms

import tf_filters as tf
from scipy import optimize

from IPython.display import Math
import IPython.display
```

```python
dt = 1e-4
tstop = 10

ge_mu, ge_sigma, Te = 20e-9, 7e-9, 7.1e-3
gi_mu, gi_sigma, Ti = 42e-9, 9e-9, 3.7e-3
tf.Gl = 1.e-6/60
tf.Cm=250e-12

# we try a simulation with ~ntwk sim values
ge = tf.ornstein_uhlenbeck(tstop, dt, ge_mu, ge_sigma, Te)
gi = tf.ornstein_uhlenbeck(tstop, dt, gi_mu, gi_sigma, Ti)
#ge = tf.white_gaussian(tstop, dt, ge_mu, ge_sigma)
#gi = tf.white_gaussian(tstop, dt, gi_mu, gi_sigma)

ge,gi = tf.rectify(ge), tf.rectify(gi) ### non zero conductances !!

t, v, spikes = tf.launch_sim(dt, ge, gi, spiking_mech=False)
```

```python
def calculate_acf(x, maxlag):
    x_nomean = x - x.mean()
    acf = np.convolve(x_nomean, x_nomean[-maxlag::-1], mode='valid')
    acf /= (len(x_nomean) - maxlag)
    acf /= np.var(x)
    return acf

exp_f = lambda t, tau: np.exp(-t/tau) 
max_lag = 1000

def fit_acf(time, ge_acf):
    params, pcov = optimize.curve_fit(exp_f, time, ge_acf)
    return params
def plot_acf(time, ge_acf, params, color='k', labelshift=0):
    ax = plt.gca()
    
    trans = transforms.blended_transform_factory(ax.transAxes,
                                                 ax.transData)
    plt.plot(time, ge_acf , marker='.', color=color)
    plt.plot(time, exp_f(time, params[0]), ls='-', color=color)
    ax.text(1.0, ge_acf[-1]+labelshift,
            '$\\tau$ = %.1f ms' % (params[0] * 1000, ),
            ha='right', size=16,
             transform=trans, clip_on=False, color=color)
    
def plot_and_fit_acf(ge, color, max_lag=max_lag, labelshift=0):
    time = np.arange(max_lag)*dt
 
    ge_acf = calculate_acf(ge, max_lag)
    
    params = fit_acf(time, ge_acf)
    plot_acf(time, ge_acf, params, color, labelshift=labelshift)
    
    return params[0]
    
```

## Synaptic conductances

```python
tau_ge = plot_and_fit_acf(ge, 'r', 100)
tau_gi = plot_and_fit_acf(gi, 'b', 100)
plt.title('conductances')
```

## Membrane potential

```python
ax = plt.subplot(111)
tau_m = plot_and_fit_acf(v, 'g', 400, labelshift=0.1)

plt.title('membrane potential')
plt.xlabel('time (ms)')
plt.ylabel('ACF (normalised)')
ax.text(0.5, 0.4, 
        ("$\\bar{g}_e=%.1e$\n$\sigma_{g_e}=%.1e$\n"+
        "$\\bar{g}_i=%.1e$\n$\sigma_{g_i}=%.1e$\n"+
        "$C_m=%.1e$\n$g_L=%.1e$") % (ge_mu, ge_sigma, gi_mu, gi_sigma, tf.Cm, tf.Gl), 
        transform=ax.transAxes, clip_on=False, size=16)
plt.tight_layout()
plt.savefig('vm_acf_white_noise_conductance.pdf')
```
