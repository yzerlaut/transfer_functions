```python
#### TESTING THE EFFECT OF GAUSSIAN WHITE NOISE ####

dt = 1e-4
tstop = 1000

# current stimulation values
I0 = 150*1e-12 # Ampere
sI = 400*1e-12 # A

# we try a simulation with ~ntwk sim values
i_gen = pg.white_gaussian(tstop, dt, I0, sI)

t, i, v, spikes = pg.launch_sim_current(dt, i_gen, spiking_mech=True,max_spikes=10000)

print " spike number : ",len(spikes)
print " firing frequency : ",1.*len(spikes)/(len(v)*dt)

plt.figure(figsize=(15,4))
plt.plot(t[:1e4]*1e3,v[:1e4]*1e3)
```

```python
def isolated_spikes(spikes, min_isi):
    isi = np.diff(spikes)
    is_short_isi  = (isi<min_isi)
    is_isolated = ~(np.r_[False, is_short_isi])
    spikes = spikes[is_isolated]
    return spikes

spikes = np.array(spikes)
iso_spikes = isolated_spikes(spikes, 75e-3)
```

```python
window = (-50*1e-3,0)
lag_i,sta_i = tf.calculate_sta(dt, i-i.mean(), iso_spikes, window)

plt.figure()
plt.plot(lag_i*1e3, sta_i,label='STA filter')
plt.xlabel('ms')
# we compare it with the membrane green function 
lag_th = np.arange(0,1000)*dt #ready in ms
green_func = np.exp(-lag_th*Gl/Cm)*(sta_i.max()-sta_i.min())
plt.plot(lag_th[::-1]*1e3-100, green_func,label='Green function')
plt.legend(loc='upper left')
plt.xlim([-50,0])
```

```python
## Non linearity 1D

i_bins, i_spike, i_total = tf.estimate_nonlinearity_1d(i, sta_i, iso_spikes[:-1], dt)
#i_bins, i_spike, i_total = tf.estimate_nonlinearity_1d(v*1e3, np.ones(len(sta_i), dtype=float)/len(sta_i), iso_spikes, dt)
plt.subplot(221); plt.plot(i_bins, i_spike)
plt.axvline(color='k')
plt.title('spike-triggered ensemble')
plt.ylabel('excitation')
plt.subplot(222); plt.plot(i_bins, i_spike*1./i_total, '.')
plt.ylim((0,1e-4))
plt.title('nonlinearity')
```

```python
def trigger_segments(stim, spikes, n, dt):
    spk_idx = (spikes/dt).astype(int)
    spk_idx = spk_idx[spk_idx>n]
    segs = np.array([stim[i-n+1:i+1] for i in spk_idx])
    return segs
```

```python
n_sta = len(sta_i)
segs = trigger_segments(i, iso_spikes, n_sta, dt)
```

```python
dsegs = segs-segs.mean(0)[None,:]
C_spike = np.dot(dsegs.T, dsegs)/dsegs.shape[0]

```

```python
from scipy import linalg
C_prior = linalg.toeplitz(np.correlate(i[:],i[:-n_sta+1]))
```

```python
DeltaC = C_spike - C_prior
evals, evecs = np.linalg.eig(DeltaC)
```
