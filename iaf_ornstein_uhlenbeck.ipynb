```python
import numpy as np
import matplotlib.pylab as plt

## PHYSICAL UNITS ###

# parameters
Gl= 1e-8 ; Cm= 0.2*1e-9; El = -65*1e-3 
Ne = 200 ; Qe = 6*1e-9 ; Te = 5*1e-3; Ee = 0*1e-3 ; fe = 4
Ni = 50 ; Qi = 64*1e-9 ; Ti = 10*1e-3; Ei = -80*1e-3 ; fi= 10
Vthre=-50*1e-3; refrac = 5*1e-3 ; 

import tf_filters as tf
```

```python
#### TESTING THE EFFECT OF GAUSSIAN WHITE NOISE ####

dt = 1e-4
tstop = 100 

# we try a simulation with ~ntwk sim values
ge = tf.ornstein_uhlenbeck(tstop, dt, 11e-9, 5e-9,Te)
gi = tf.ornstein_uhlenbeck(tstop, dt, 15e-9, 7e-9,Ti)
ge,gi = tf.rectify(ge), tf.rectify(gi) ### non zero conductances !!

t, v, spikes = tf.launch_sim(dt, ge, gi)
print " spike number : ",len(spikes)
print " firing frequency : ",1.*len(spikes)/tstop

plt.figure(figsize=(15,4))
plt.plot(t[:1e4]*1e3,v[:1e4]*1e3)

window = (-20*1e-3,0)

lag_ge,sta_ge = tf.calculate_sta(dt, ge, spikes, window)
plt.figure()
plt.plot(lag_ge*1e3, sta_ge);plt.xlabel('ms')
lag_gi,sta_gi = tf.calculate_sta(dt, gi, spikes, window)
plt.plot(lag_gi*1e3, sta_gi);plt.xlabel('ms')
```

```python
## TESTING THE EFFECT OF GAUSSIAN WHITE NOISE ####
# DOES THE FILTERS CHANGE WITH THE WN PARAMETERS ? ... #

dt = 1e-4
tstop = 1

conductance_params = [
    dict(ge_mu=11e-9, ge_sigma=5e-9, gi_mu=16e-9, gi_sigma=7e-9), 
    dict(ge_mu=12e-9, ge_sigma=5e-9, gi_mu=16e-9, gi_sigma=7e-9), 
    dict(ge_mu=11e-9, ge_sigma=5e-9, gi_mu=15e-9, gi_sigma=7e-9)
]

plt.figure(figsize=(13,4))
    
for params in conductance_params:

    
    ge = tf.ornstein_uhlenbeck(tstop, dt, params['ge_mu'], params['ge_sigma'],Te)
    gi = tf.ornstein_uhlenbeck(tstop, dt, params['gi_mu'], params['gi_sigma'],Ti)
    
    ge,gi = tf.rectify(ge), tf.rectify(gi) ### non zero conductances !!
    
    t, v, spikes = tf.launch_sim(dt, ge, gi)
    print " spike number : ",len(spikes)," -> firing frequency : ",1.*len(spikes)/tstop
       
    window = (-10*1e-3,0)
    
    lag_ge,sta_ge = tf.calculate_sta(dt, ge, spikes, window)
    lag_gi,sta_gi = tf.calculate_sta(dt, gi, spikes, window)
    
    plt.subplot(121)
    plt.plot(lag_ge*1e3, sta_ge);plt.xlabel('ms')
    plt.subplot(122)
    plt.plot(lag_gi*1e3, sta_gi);plt.xlabel('ms')
```
