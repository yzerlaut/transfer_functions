```python
import process_generators as pg

Gl= 1e-8 ; Cm= 0.2*1e-9; El = -65*1e-3 
Ne = 200 ; Qe = 6*1e-9 ; Te = 5*1e-3; Ee = 0*1e-3 ; fe = 4
Ni = 50 ; Qi = 64*1e-9 ; Ti = 10*1e-3; Ei = -80*1e-3 ; fi= 10
Vthre=-50*1e-3; refrac = 5*1e-3 ; 

mu_exc, mu_inh = Qe*Te*Ne*fe, Qi*Ti*Ni*fi
sigma_exc, sigma_inh = Qe*np.sqrt(Te*fe*Ne), Qi*np.sqrt(Ti*fi*Ni)

E_exc, E_inh = 0, -80e-3
dt = 0.1e-3
tau_m = 0.010


def ornstein_uhlenbeck_current(dt, mu, sigma, tau, store=True):
    """return a realisation of ornstein-uhlenbeck process"""

    diffcoef = 2*sigma**2/tau
    A = np.sqrt(diffcoef*tau/2.*(1-np.exp(-2*dt/tau)))
    y0 = mu
    
    past_values = []
    
    def _calc(y, t):
        noise = np.random.randn()
        y = y0 + (y - y0) * np.exp(-dt/tau) + A * noise
        return y
    
    def _store(y, t):
        y = _calc(y, t)
        past_values.append(y)
        return y
        
    if store:
        return _store, past_values
    else:
        return _calc
        
def reset(tmax, dt, i_exc, v_thre, v_reset):
    max_steps = int(tmax/dt)
    spikes, time = [], []
    v = v_reset
    for i in xrange(max_steps):
        t = i*dt
        v = i_exc(v, t)
        time.append(t)
        if v>v_thre:
            spikes.append(t)
            v = v_reset
    return time, spikes
            
        
def integrate_reset(tmax, dt, i_exc, v_thre, v_reset, tau_m, store=True):
    max_steps = int(tmax/dt)
    spikes, time, mem_potential = [], [], []
    v = v_reset
    for i in xrange(max_steps):
        t = i*dt
        v = -v*dt/tau_m + i_exc(v, t)
        time.append(t)
        if v>v_thre:
            spikes.append(t)
            v = v_reset
        if store:
            mem_potential.append(v)
    return time, spikes, mem_potential

```

```python
def calc_fr(tmax, dt, mu, sigma, tau, threshold):
    gexc = ornstein_uhlenbeck_current(dt, mu, sigma, tau, store=False)
    time, spikes, _ = integrate_reset(tmax, dt, gexc, threshold, mu, store=False)
    return len(spikes)/tmax

param, fr = np.array([(_, calc_fr(tmax, _, mu_exc, sigma_exc, Te, v_thresh))
                      for _ in np.logspace(-1, -5, 10)]).T

plt.semilogx(param, fr)
```

```python
%%cython
cimport cython
from libc.math cimport exp, sqrt, pow, log, erf, cos
from libc.stdlib cimport rand, RAND_MAX

cdef double PI = 3.14159265358979323846

@cython.cdivision(True)
cdef double drand():
    return (rand()+1.0)/(RAND_MAX+1.0);

@cython.cdivision(True)
cdef double random_normal():
    return sqrt(-2*log(drand())) * cos(2*PI*drand());


@cython.cdivision(True)
cdef double ou_sample(double y, double dt, double mu, double sigma, double tau):
    cdef double diffcoef = 2*sigma**2/tau
    cdef double A = sqrt(diffcoef*tau/2.*(1-exp(-2*dt/tau)))
    cdef double noise = random_normal()
    y = mu + (y - mu) * exp(-dt/tau) + A * noise
    return y

@cython.cdivision(True)
def fast_integrate_reset(double tmax, double dt, double v_thre, double v_reset, double tau, 
                    double imu, double isigma, double itau):
    cdef int max_steps = (int)(tmax/dt)
    spikes, mem_potential = [], []
    cdef double v = v_reset
    cdef double t
    cdef int i
    for i in range(max_steps):
        t = i*dt
        v = -v*dt/tau + ou_sample(v, dt, imu, isigma, itau)
        if v>v_thre:
            spikes.append(t)
            v = v_reset
        #mem_potential.append(v)
    return spikes, mem_potential
```
