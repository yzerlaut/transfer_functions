```python
import numpy as np
import matplotlib.pylab as plt

## PHYSICAL UNITS ###
dt = 1e-4
tstop = 11 # we will discard the 1 second for security (but way too much)

# parameters
Gl= 1e-8 ; Cm= 0.2*1e-9; El = -65*1e-3 
Ne = 200 ; Qe = 6*1e-9 ; Te = 5*1e-3; Ee = 0*1e-3 ; fe =3.4
Ni = 50 ; Qi = 64*1e-9 ; Ti = 10*1e-3; Ei = -80*1e-3 ; fi=1
Vthre=-50*1e-3; refrac = 5*1e-3 ; 
```

```python
def ornstein_uhlenbeck(mu, sigma, tau, tmax=1000, dt=1.):
    """return a realisation of ornstein-uhlenbeck process"""
    diffcoef = 2*sigma**2
    y0 = mu
    n_steps = int(tmax/dt)
    A = np.sqrt(diffcoef*tau/2.*(1-np.exp(-2*dt/tau)))
    noise = np.random.randn(n_steps)
    y = np.zeros(n_steps)
    y[0] = y0
    for i in range(n_steps-1):
        y[i+1] = y0 + (y[i]-y0)*np.exp(-dt/tau)+A*noise[i]
    return y


def launch_sim(mge,sge,mgi,sgi,spiking_mech=True,past_sta=20*1e-3):
    # we generate the Ornstein-Uhlenbeck Process
    ge = ornstein_uhlenbeck(mge,sge,Te,tstop,dt=dt)
    gi = ornstein_uhlenbeck(mgi,sgi,Ti,tstop,dt=dt)
    # we insure that it does not take negative values !!!
    np.where(ge<0,ge,0);np.where(gi<0,gi,0)

    t = np.arange(0,tstop,dt)
    v = np.ones(t.size)*El # refractory if not changed

    last_spike = -tstop # for the refractory period
    spikes = []
    ge_sta = [];gi_sta = [] # for sta analysis
    for i in range(t.size-1):
        if (t[i+1]-last_spike)>refrac:
            v[i+1] = v[i] + dt/Cm*( Gl*(El-v[i]) + gi[i]*(Ei-v[i]) + ge[i]*(Ee-v[i]) )
        if v[i+1]>Vthre and spiking_mech:
            last_spike = t[i+1]
            spikes.append(last_spike)
            #v[i+1]=0 ## UNCOMMENT to have a nice spike shape...
            if t[i+1]>past_sta:
                ge_sta.append(ge[i-int(past_sta/dt):i])
                gi_sta.append(gi[i-int(past_sta/dt):i])
    # we restrict our analysis to the stationary behaviour, no transients
    v = v[int(tstop/10./dt):];t=t[int(tstop/10./dt):]
    ge=ge[int(tstop/10./dt):];gi=gi[int(tstop/10./dt):]
    return t,v,ge,gi,spikes,ge_sta,gi_sta
```

```python
# we try a simulation with ~ntwk sim values
mge,sge = Ne*Qe*fe*Te,Qe*np.sqrt(fe*Te*Ne/2)
mgi,sgi = Ni*Qi*fi*Ti,Qi*np.sqrt(fi*Ti*Ni/2)
t,v,ge,gi,spikes,ge_sta,gi_sta = launch_sim(mge,sge,mgi,sgi)
plt.figure(figsize=(15,4))
plt.plot(t,v*1e3)
```

```python
def gaussian_1d(x, mu, sigma):
    return 1./(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2*sigma**2))

plt.figure(figsize=(10,4))

plt.subplot(121)
n, bins = np.histogram(1e9*ge,100,density=True)
plt.bar(bins[:-1],n,width=bins[1]-bins[0], label='simulation')
th_ge_density = gaussian_1d(bins,1e9*mge,1e9*sge)
th_ge_density /= th_ge_density.sum()*(bins[1]-bins[0])
plt.plot(bins,th_ge_density,'k--',lw=3 ,label='theory')
plt.legend(frameon=False,loc='upper left')
plt.xlabel('ge (nS)')

plt.subplot(122)
n, bins = np.histogram(1e9*gi,100,density=True)
plt.bar(bins[:-1],n,color='r',width=bins[1]-bins[0],label='simulation')
plt.plot(bins, gaussian_1d(bins,1e9*mgi,1e9*sgi),'k--',lw=3 ,label='theory')
plt.legend(frameon=False,loc='upper left')
plt.xlabel('gi (nS)')
```

```python
## study of the Ornstein-Uhlenbeck process
vec_change = np.linspace(1,5,10)
mean = 10
std = 1
T = 1
var_effect = [];T_effect = [];dt_effect=[]
for change in vec_change:
    var_effect.append(ornstein_uhlenbeck(mean,std*change,T,tmax=1000,dt=.1).std())
    T_effect.append(ornstein_uhlenbeck(mean,std,T*change,tmax=1000,dt=.1).std())
    dt_effect.append(ornstein_uhlenbeck(mean,std,T,tmax=1000,dt=.1*change).std())

plt.figure(figsize=(16,4))
plt.subplot(131)
plt.title('effect of the variance change')
plt.plot(std*vec_change,var_effect,'bo',lw=0,label='measured value')
plt.plot(std*vec_change,std*vec_change,'k',label='expected value')
plt.legend(frameon=False,loc='upper left')
plt.subplot(132)
plt.title('effect of the time constant change')
plt.plot(std*vec_change,T_effect,'bo',lw=0,label='measured value')
plt.plot(std*vec_change,std*vec_change,'k',label='expected value')
plt.legend(frameon=False,loc='upper left')
plt.subplot(133)
plt.title('effect of the time step (dt)')
plt.plot(std*vec_change,dt_effect,'bo',lw=0,label='measured value')
plt.plot(std*vec_change,std+0*vec_change,'k',lw=2,label='expected value')
plt.legend(frameon=False,loc='upper left')
```

```python
def autocorrel(signal,t):
    """ argument : signal (np.array), t (np.array) -> returns : autocorrel (np.array), time_shift (np.array) """
    signal = (signal-signal.mean())/signal.std()
    cr = np.correlate(signal,signal,"full")
    time_shift = np.concatenate((-t[::-1],t[1:]))
    return cr/cr.max(),time_shift

plt.figure(figsize=(12,4))

plt.subplot(121)
ac_ge,t_shift = autocorrel(ge,t)
plt.plot(t_shift,ac_ge);plt.xlabel('time shift (ms)')

plt.subplot(122)
ac_gi,t_shift = autocorrel(gi,t)
plt.plot(t_shift,ac_gi,'r');plt.xlabel('time shift (ms)')
```

```python
def rudolph_rho_v(V,ge,sge,gi,sgi):
    """ here the membrane parameters are global variablaes """

    A1 = - (2*Cm*(ge+gi)+2*Cm*Gl+sge**2*Te+sgi**2*Ti)
    A1 /= 2*(sge**2*Te+sgi**2*Ti)
    A2 = 2*Cm*(Gl*(sge**2*Te*(El-Ee)+sgi**2*Ti*(El-Ei))+\
               (ge*sgi**2*Ti-gi*sge**2*Te)*(Ee-Ei))
    A2 /= (Ee-Ei)*np.sqrt(sge**2*Te*sgi**2*Ti)*(sge**2*Te+sgi**2*Ti)
    a_ln = sge**2*Te/Cm*(V-Ee)**2+sgi**2*Ti/Cm*(V-Ei)**2
    a_arctan = sge**2*Te*(V-Ee)+sgi**2*Ti*(V-Ei)
    a_arctan /= (Ee-Ei)*np.sqrt(sge**2*Te*sgi**2*Ti)
    return np.exp(A1*np.log(a_ln)+A2*np.arctan(a_arctan))

vv = np.linspace(-90*1e-3,20*1e-3,2e2)
th_rudolph = rudolph_rho_v(vv,mge,sge,mgi,sgi)
th_rudolph /= np.sum(th_rudolph)*(vv[1]-vv[0])
n, bins = np.histogram(v,bins=vv,density=True)
plt.bar(bins[:-1],n,width=bins[1]-bins[0],label='simulation')
#plt.plot(vv,th_rudolph,'k--',lw=3,label='theory')
plt.title("distribution of projection coefficients")
plt.legend(frameon=False,loc='upper left')
```
