```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import transforms

import tf_filters as tf
from scipy import optimize

from IPython.display import Math
import IPython.display
```

```python
dt = 1e-4
tstop = 10

ge_mu, ge_sigma, Te = 11e-9,5e-9, 5e-3
gi_mu, gi_sigma, Ti = 15e-9,7e-9, 10e-3

# we try a simulation with ~ntwk sim values
ge = tf.ornstein_uhlenbeck(tstop, dt, ge_mu, ge_sigma, Te)
gi = tf.ornstein_uhlenbeck(tstop, dt, gi_mu, gi_sigma, Ti)
ge,gi = tf.rectify(ge), tf.rectify(gi) ### non zero conductances !!

t, v, spikes = tf.launch_sim(dt, ge, gi, spiking_mech=False)
```

```python
def calculate_acf(x, maxlag):
    x_nomean = x - x.mean()
    acf = np.convolve(x_nomean, x_nomean[-maxlag::-1], mode='valid')
    acf /= (len(x_nomean) - maxlag)
    acf /= np.var(x)
    return acf

exp_f = lambda t, tau: np.exp(-t/tau) 
max_lag = 1000


def plot_and_fit_acf(ge, color, max_lag=max_lag):
    time = np.arange(max_lag)*dt
    ax = plt.gca()
    trans = transforms.blended_transform_factory(ax.transAxes,
                                                 ax.transData)
    ge_acf = calculate_acf(ge, max_lag)
    plt.plot(time, ge_acf , marker='.', color=color)
    params, pcov = optimize.curve_fit(exp_f, time, ge_acf)
    plt.plot(time, exp_f(time, params[0]), ls='-', color=color)
    ax.text(1.015, ge_acf[-1],
            '$\\tau$ = %.1f ms' % (params[0] * 1000, ),
             transform=trans, clip_on=False, color=color)
    return params[0]
    
```

## Synaptic conductances

```python
tau_ge = plot_and_fit_acf(ge, 'r', 100)
tau_gi = plot_and_fit_acf(gi, 'b', 100)
plt.title('conductances')
```

## Membrane potential

```python
tau_m = plot_and_fit_acf(v, 'g', 200)
plt.title('membrane potential')
```

##Synaptic currents

```python
i_e = ge * (tf.Ee - v)
i_i = gi * (tf.Ei - v)
i_total = i_e + i_i
tau_ie = plot_and_fit_acf(i_e, 'r', 100)
tau_ii = plot_and_fit_acf(i_i, 'b', 100)
plot_and_fit_acf(i_total, 'g', 100)
plt.title('currents')
```

## Time constants summary

```python
Tm = tf.Cm/tf.Gl
print "Membrane: %.1f ms (actual), %.1f ms (analytical)" % (tau_m*1e3, Tm*1e3)

tau_ie_est = 1./((tau_ge+Tm)/(tau_ge*Tm))
tau_ii_est = 1./((tau_gi+Tm)/(tau_gi*Tm))
print "g_e: %.1f ms " % (tau_ge*1e3, ), "g_i: %.1f ms " % (tau_gi*1e3, )
print "I_e: %.1f ms (actual), %.1f ms (analytical)" % (tau_ie*1e3, tau_ie_est*1e3)
print "I_i: %.1f ms (actual), %.1f ms (analytical)" % (tau_ii*1e3, tau_ii_est*1e3)
```

## Synaptic currents (mean membrane potential)

```python
im_e = ge * (tf.Ee - v.mean())
im_i = gi * (tf.Ei - v.mean())
im_total = im_e + im_i
plot_and_fit_acf(im_e, 'r', 100)
plot_and_fit_acf(im_i, 'b', 100)
plot_and_fit_acf(im_total, 'g', 100)
plt.title('currents')
```

```python
t, v_current, spikes = tf.launch_sim_current(dt, i_total, spiking_mech=False)
plot_and_fit_acf(v_current, 'g', 2500)
plt.title('membrane potential (current-based)')
```
