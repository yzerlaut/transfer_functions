```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
isi = np.exp(-np.random.rand(10))
spikes = np.cumsum(isi)
```

```python
bins = np.linspace(0, spikes.max(), 100)
```

```python
def exp_heaviside(x):
    return np.where(x>0, np.exp(-x), 0)
tau = 0.2
g_trace = exp_heaviside((bins[:, None]-spikes[None, :])/tau).sum(1)
```

```python
plt.plot(bins, g_trace)
plt.vlines(spikes, 0, 1)
```
