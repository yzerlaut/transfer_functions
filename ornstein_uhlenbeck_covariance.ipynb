```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg, special
```

```python
mu, theta, sigma = 1, .3,1
```

```python
def cov_matrix(sigma, theta, tmax, dt=1):
    """calculate and return covariance matrix of ornstein-uhlenbeck proccess"""
    time_lag = np.arange(int(tmax/dt))*dt
    column = sigma**2/(2.*theta)*np.exp(-theta*time_lag)
    cov = linalg.toeplitz(column)
    return cov

def rotate_matrix(vec, mat):
    return np.dot(np.dot(vec, mat), vec.T)

def gaussian_1d(x, mu, sigma):
    return 1./(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2.*sigma**2))

def prob_threshold_crossing(theta, mu, sigma):
    return special.erfc((theta-mu)/(np.sqrt(2)*sigma))/2.

def calculate_cov_matrix(y, tmax, dt=1.):
    y_nomean = y - np.mean(y)
    n_max = int(tmax/dt)
    n_pts = len(y_nomean)
    cov = [np.mean(y_nomean[i:]*y_nomean[:n_pts-i])
          for i in range(n_max)]
    return linalg.toeplitz(cov)
    
```

```python
def ornstein_uhlenbeck(mu, sigma, theta, tmax, dt=1.):
    """return a realisation of ornstein-uhlenbeck process"""

    diffcoef = sigma**2
    y0 = mu
    tau = 1./theta
    
    n_steps = int(tmax/dt)
    
    A = np.sqrt(diffcoef*tau/2.*(1-np.exp(-2*dt/tau)))

    noise = np.random.randn(n_steps)

    y = np.zeros(n_steps)
    y[0] = y0

    for i in range(n_steps-1):
        y[i+1] = y0 + (y[i]-y0)*np.exp(-dt/tau)+A*noise[i]
    return y
```

Simulate spikes
---------------

```python
filter_length = 10
time = np.arange(-filter_length, 0, 1)
G_filter = np.exp(time/2.)
#normalize to unit variance
G_filter = G_filter/np.sqrt(np.sum(G_filter**2))
plt.plot(time, G_filter)
plt.xlabel('time lag')
plt.title('linear filter')
```

```python
threshold = 3.
projected_samples = np.convolve(ou_samples, G_filter)
spikes = (projected_samples>threshold)
spike_rate = np.sum(spikes)*1./len(spikes)
plt.plot(spikes[:1000])
plt.title('simulated spikes')
```

Estimate firing rate
--------------------

```python
cov_theory = cov_matrix(sigma, theta, filter_length)
sigma_proj = np.sqrt(rotate_matrix(G_filter[::-1], cov_theory))
mu_proj = np.sum(G_filter)*mu
prob_theory = prob_threshold_crossing(threshold, mu_proj, sigma_proj)
print "Spike rate (estimated/theory):", spike_rate,"/", prob_theory
print "Projection std. (estimated/theory):", np.std(projected_samples),"/", sigma_proj
```

```python
n, bins = np.histogram(projected_samples,100, density=True)
plt.bar(bins[:-1],n,width=bins[1]-bins[0], label='simulation',fc='none')
plt.plot(bins, gaussian_1d(bins,mu_proj, sigma_proj),'r', label='theory')
plt.title("distribution of projection coefficients")
plt.axvline(threshold,label='threshold')
plt.legend(frameon=False,loc='upper left')
```

```python
def compare_theory_simulation(mu, theta, sigma, threshold):
    npts = 50000
    ou_samples = ornstein_uhlenbeck(mu, sigma, theta, npts)
    spikes = np.convolve(ou_samples, G_filter)>threshold
    spike_rate = np.sum(spikes)*1./len(spikes)
    
    cov_theory = cov_matrix(sigma, theta, filter_length)
    sigma_proj = np.sqrt(rotate_matrix(G_filter[::-1], cov_theory))
    mu_proj = np.sum(G_filter)*mu
    prob_theory = prob_threshold_crossing(threshold, mu_proj, sigma_proj)
    
    return spike_rate, prob_theory

def plot_comparison(value_list, title=None, labels=True):
    simulation, theory = np.array(value_list).T
    plt.plot(simulation, theory, 'o')
    xmin, xmax = plt.xlim()
    ymin, ymax = plt.ylim()
    mmax = max(xmax, ymax)
    mmin = min(xmin, ymin)
    plt.plot([mmin, mmax], [mmin, mmax], 'k-')
    plt.xticks([mmin, mmax])
    plt.yticks([mmin, mmax])
    if labels:
        plt.xlabel('spike rate - simulation',labelpad=-8)
        plt.ylabel('spike rate - theory', labelpad=-8)
        
    ax = plt.gca()
    if title:
        plt.text(0.1, 0.8, title, transform=ax.transAxes)
    plt.xlim([mmin, mmax])
    plt.ylim([mmin, mmax])
```

### Compare theoretical and estimated firing rates

```python
vary_theta = [compare_theory_simulation(mu, _, sigma, threshold) for _ in np.linspace(0.5,2,5)]
vary_sigma = [compare_theory_simulation(mu, theta, _, threshold) for _ in np.linspace(0.5,5,5)]
vary_mu = [compare_theory_simulation(_, theta, sigma, threshold) for _ in np.linspace(0.5,2,5)]
vary_threshold = [compare_theory_simulation(mu, theta, sigma, _) for _ in np.linspace(0.5,2,5)]
```

```python
plt.figure(figsize=(8,5))
plt.subplot(221)
plot_comparison(vary_theta,'vary theta')
plt.subplot(222)
plot_comparison(vary_sigma,'vary sigma', labels=False)
plt.subplot(223)
plot_comparison(vary_mu, 'vary mu', labels=False)
plt.subplot(224)
plot_comparison(vary_threshold, 'vary threshold', labels=False)
```

Recover filters with STA
------------------------

Compute the spike-triggered average and whiten it using the whitening transformation $z=Vx$, where $V=E D^{-\frac{1}{2}} E^T$. ($E$ is the matrix of eigenvectors aranged in columns and $D$ is the diagonal matrix of eigenvalues of covariance matrix of $x$)

```python
sta = np.correlate(ou_samples[:]-ou_samples.mean(), spikes[:])[::-1]
sta = sta/np.sum(spikes)

evals, evecs = np.linalg.eig(cov_theory)
d = 1./np.sqrt(evals)
whitening_filter = np.dot(evecs,np.dot(np.diag(d),evecs.T))
sta_whitened = np.dot(whitening_filter, sta)


plt.plot(time, sta_whitened, label='whitened STA')
plt.plot(time, G_filter, label='original STA')
plt.plot(time, sta, label='raw STA')
plt.title('filter estimates')
plt.xlabel('time lag')
plt.legend(loc='upper left', frameon=False)
```

```python

def gaussian_nd(x, mu, cxx):
    d=len(x)
    norm = np.sqrt((2*np.pi)**(d)*np.linalg.det(cxx))
    
    inv_cxx = np.linalg.inv(cxx)
    delta_x  = x - mu
    exponent = -1/2.*np.dot(np.dot(delta_x, inv_cxx), delta_x)
    return 1./norm*np.exp(exponent)

print gaussian_nd(np.array([0]), np.array([1]), np.array([[1]])), gaussian_1d(0, 1., 1.)
print gaussian_nd(np.array([1]), np.array([0]), np.array([[3]])), gaussian_1d(1, 0, np.sqrt(3))
print gaussian_nd(np.array([1]), np.array([3]), np.array([[3]])), gaussian_1d(1, 3, np.sqrt(3))
```

```python
from scipy import integrate

def ou_rate(mu, sigma, tau, threshold):
    cxx = cov_matrix(sigma*np.sqrt(2./tau), 1./tau, 2*dt, dt)
    prob = lambda x, y: gaussian_nd(np.array([x,y]), mu, cxx)

    integrate.dblquad?
    gfun = lambda x: threshold
    hfun = lambda x: np.inf
    p, err = integrate.dblquad(prob, -np.inf, threshold, gfun, hfun)
    return p/dt
print ou_rate(mu, sigma, tau, threshold), calc_fr(mu, sigma, tau, threshold)
```
