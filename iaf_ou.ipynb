```python
import numpy as np
import matplotlib.pylab as plt

## PHYSICAL UNITS ###
dt = 1e-4
tstop = 11 # we will discard the 1 second for security (but way too much)

# parameters
Gl= 1e-8 ; Cm= 0.2*1e-9; El = -65*1e-3 
Ne = 200 ; Qe = 6*1e-9 ; Te = 5*1e-3; Ee = 0*1e-3 ; fe = 4
Ni = 50 ; Qi = 64*1e-9 ; Ti = 10*1e-3; Ei = -80*1e-3 ; fi= 10
Vthre=-50*1e-3; refrac = 5*1e-3 ; 
```

```python
def ornstein_uhlenbeck(tmax, dt, mu, sigma, tau):
    """return a realisation of ornstein-uhlenbeck process"""

    diffcoef = 2*sigma**2/tau
    y0 = mu
    n_steps = int(tmax/dt)
    A = np.sqrt(diffcoef*tau/2.*(1-np.exp(-2*dt/tau)))
    noise = np.random.randn(n_steps)
    y = np.zeros(n_steps)
    y[0] = y0
    for i in range(n_steps-1):
        y[i+1] = y0 + (y[i]-y0)*np.exp(-dt/tau)+A*noise[i]
    return y

def white_gaussian(tmax, dt, mu, sigma):
    """return a realisation of a gaussian white noise process"""
    
    return np.random.randn(int(tmax/dt)) * sigma + mu

def rectify(x):
    """ take an array and repaces its negative points and replace it with 0 values """
    out = x.copy()
    out[out<0]=0
    return out


def launch_sim(dt, ge, gi, spiking_mech=True):
    """ functions that solve the membrane equations for 2 time varying 
    excitatory and inhibitory conductances
    N.B. reversal potentials, membrane prop. should be global """
    
    try: 
        tstop = len(ge)*dt
    except TypeError:
        pass
    try: 
        tstop = len(gi)*dt
    except TypeError:
        pass
    
    t = np.arange(0,tstop,dt)
    v = np.ones(t.size)*El # refractory if not changed

    last_spike = -tstop # time of the last spike, for the refractory period
    spikes = []
    for i in range(t.size-1):
        if (t[i+1]-last_spike)>refrac:
            v[i+1] = v[i] + dt/Cm*( Gl*(El-v[i]) + gi[i]*(Ei-v[i]) + ge[i]*(Ee-v[i]) )
        if v[i+1]>Vthre and spiking_mech:
            last_spike = t[i+1]
            spikes.append(last_spike)
            #v[i+1]=0 ## UNCOMMENT to have a nice spike shape...
    return t, v, spikes

def calculate_sta(dt, x, spikes, window):
    """ function that calculate the STA of x 
    arguments : dt (float), x, spikes (np.array), window (tuple of minimum and maximum lag in time units)"""
    
    spike_train = np.histogram(spikes, np.arange(len(v))*dt)[0]
    min_index = int(window[0]/dt)
    max_index = len(spike_train) - int(window[1]/dt)
    sta = np.correlate(x, spike_train[-min_index-1:max_index]) ### TO BE CHECK
    lag = np.arange(len(sta))*dt+window[0]
    return lag,sta/len(spikes)
    
    
```

```python
#### TESTING THE EFFECT OF GAUSSIAN WHITE NOISE ####

dt = 1e-4
tstop = 100 

# we try a simulation with ~ntwk sim values
ge = white_gaussian(tstop, dt, 11e-9, 5e-9)
gi = white_gaussian(tstop, dt, 15e-9, 7e-9)
ge,gi = rectify(ge), rectify(gi) ### non zero conductances !!

t, v, spikes = launch_sim(dt, ge, gi)
print " spike number : ",len(spikes)
print " firing frequency : ",1.*len(spikes)/tstop

plt.figure(figsize=(15,4))
plt.plot(t[:1e4]*1e3,v[:1e4]*1e3)

window = (-20*1e-3,0)

lag_ge,sta_ge = calculate_sta(dt, ge, spikes, window)
plt.figure()
plt.plot(lag_ge*1e3, sta_ge);plt.xlabel('ms')
lag_gi,sta_gi = calculate_sta(dt, gi, spikes, window)
plt.plot(lag_gi*1e3, sta_gi);plt.xlabel('ms')
```

```python
#### TESTING THE EFFECT OF ORNSTEIN UHLENBECK NOISE ####

dt = 1e-4
tstop = 10 

# we try a simulation with ~ntwk sim values
mge,sge = Ne*Qe*fe*Te,Qe*np.sqrt(fe*Te*Ne/2)
mgi,sgi = Ni*Qi*fi*Ti,Qi*np.sqrt(fi*Ti*Ni/2)
ge = ornstein_uhlenbeck(tstop, dt, mge, sge, Te)
gi = ornstein_uhlenbeck(tstop, dt, mgi, sgi, Ti)

t, v, spikes = launch_sim(dt, ge, gi)
```

```python
def gaussian_1d(x, mu, sigma):
    return 1./(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2*sigma**2))

plt.figure(figsize=(10,4))

plt.subplot(121)
n, bins = np.histogram(1e9*ge,100,density=True)
plt.bar(bins[:-1],n,width=bins[1]-bins[0], label='simulation')
th_ge_density = gaussian_1d(bins,1e9*mge,1e9*sge)
th_ge_density /= th_ge_density.sum()*(bins[1]-bins[0])
plt.plot(bins,th_ge_density,'k--',lw=3 ,label='theory')
plt.legend(frameon=False,loc='upper left')
plt.xlabel('ge (nS)')

plt.subplot(122)
n, bins = np.histogram(1e9*gi,100,density=True)
plt.bar(bins[:-1],n,color='r',width=bins[1]-bins[0],label='simulation')
plt.plot(bins, gaussian_1d(bins,1e9*mgi,1e9*sgi),'k--',lw=3 ,label='theory')
plt.legend(frameon=False,loc='upper left')
plt.xlabel('gi (nS)')
```

```python
def rudolph_rho_v(V,ge,sge,gi,sgi):
    """ here the membrane parameters are global variablaes """

    A1 = - (2*Cm*(ge+gi)+2*Cm*Gl+sge**2*Te+sgi**2*Ti)
    A1 /= 2*(sge**2*Te+sgi**2*Ti)
    A2 = 2*Cm*(Gl*(sge**2*Te*(El-Ee)+sgi**2*Ti*(El-Ei))+\
               (ge*sgi**2*Ti-gi*sge**2*Te)*(Ee-Ei))
    A2 /= (Ee-Ei)*np.sqrt(sge**2*Te*sgi**2*Ti)*(sge**2*Te+sgi**2*Ti)
    a_ln = sge**2*Te/Cm*(V-Ee)**2+sgi**2*Ti/Cm*(V-Ei)**2
    a_arctan = sge**2*Te*(V-Ee)+sgi**2*Ti*(V-Ei)
    a_arctan /= (Ee-Ei)*np.sqrt(sge**2*Te*sgi**2*Ti)
    return np.exp(A1*np.log(a_ln)+A2*np.arctan(a_arctan))

vv = np.linspace(-90*1e-3,20*1e-3,2e2)
th_rudolph = rudolph_rho_v(vv,mge,sge,mgi,sgi)
th_rudolph /= np.sum(th_rudolph)*(vv[1]-vv[0])
n, bins = np.histogram(v,bins=100,density=True)
plt.bar(bins[:-1],n,width=bins[1]-bins[0],label='simulation')
plt.plot(vv,th_rudolph,'k--',lw=3,label='theory')
plt.title("distribution of projection coefficients")
plt.legend(frameon=False,loc='upper left');
plt.ylim([0,60])
```
