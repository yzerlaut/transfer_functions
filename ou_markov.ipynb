```python
from scipy import special, integrate

def gauss_integral(x, mu, sigma):
    """integral of normal distribution from x to inf"""
    return special.erfc((x-mu)/np.sqrt(2)/sigma)/2.

def gaussian_1d(x, mu, sigma):
    return 1./(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2.*sigma**2))

def erf_integral(x, mu, sigma):
    """integral from -inf to x of gauss_integral"""
    x_ = (x-mu)/sigma/np.sqrt(2)
    return (x_*(1+special.erf(x_))+np.exp(-x_**2)/np.sqrt(np.pi))*sigma*np.sqrt(2)

def ou_rate_markov(mu, sigma, tau, threshold, dt=0.00001):
   
    mu_ou = mu*dt/tau
    sigma_ou = sigma*(dt/tau)
    integrand = lambda x: (gaussian_1d(x, mu, sigma)*(gauss_integral(threshold, x*(1-dt/tau)+mu_ou, sigma_ou)))
    prob = integrate.quad(integrand, -np.inf, threshold)[0]
    return prob/dt

def ou_rate_markov_anal(mu, sigma, tau, threshold):
   
    stationary = gaussian_1d(threshold, mu, sigma)
    markov_integral = erf_integral(0, (threshold-mu)/tau, sigma/tau)/2.
    prob = markov_integral*stationary
    return prob

ou_rate_markov_num = np.vectorize(ou_rate_markov, otypes=[np.float64])
```

```python
mu, sigma, tau, threshold = 2, 6., 0.01, 8.
sigmas = np.linspace(1., 20., 10)
plt.subplots_adjust(wspace=0.35, hspace=0.35)
plt.subplot(2,2,1)
plt.plot(sigmas, ou_rate_markov_anal(mu, sigmas, tau, threshold), label='analytical')
plt.plot(sigmas, ou_rate_markov_num(mu, sigmas, tau, threshold), 'r.', label='numerical')
plt.axvline(sigma, color='k')
plt.xlabel('$\sigma$')
plt.legend(frameon=False, loc='upper left')

plt.subplot(2,2,2)
thresholds = np.linspace(2., 10, 10)
plt.plot(thresholds-mu, ou_rate_markov_anal(mu, sigma, tau, thresholds))
plt.plot(thresholds-mu, ou_rate_markov_num(mu, sigma, tau, thresholds), 'r.')
plt.axvline(threshold-mu, color='k')
plt.xlabel(r'$\theta-\mu$')

mus = np.linspace(0, 8, 10)
plt.plot(threshold-mus, ou_rate_markov_anal(mus, sigma, tau, threshold))

plt.subplot(2,2,3)
taus = np.logspace(-3, np.log10(0.019), 10)
plt.plot(taus*1000, ou_rate_markov_anal(mu, sigma, taus, threshold))
plt.plot(taus*1000, ou_rate_markov_num(mu, sigma, taus, threshold), 'r.')

plt.axvline(tau*1000, color='k')
plt.xlabel(r'$\tau$')
```
