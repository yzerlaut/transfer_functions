```python
dt = 1e-4
tstop = 50

ge_mu, ge_sigma = 11e-9,5e-9
gi_mu, gi_sigma = 17e-9,5e-9

# we try a simulation with ~ntwk sim values
ge = tf.ornstein_uhlenbeck(tstop, dt, ge_mu, ge_sigma, Te)
gi = tf.ornstein_uhlenbeck(tstop, dt, gi_mu, gi_sigma, Ti)
ge,gi = tf.rectify(ge), tf.rectify(gi) ### non zero conductances !!

t, v, spikes = tf.launch_sim(dt, ge, gi, max_spikes = 200)
print " spike number : ",len(spikes)
print " firing frequency : ",1.*len(spikes)/tstop

plt.figure(figsize=(15,4))
plt.plot(t[:1e4]*1e3,v[:1e4]*1e3)

window = (-30*1e-3,0)

lag_ge,sta_ge = tf.calculate_sta(dt, ge, spikes, window)
cov_ge = tf.estimate_cov_matrix(ge, len(sta_ge))
sta_filtered_ge = tf.whiten_filter(cov_ge,sta_ge)
th_cov_ge = tf.ou_cov_matrix(ge_sigma, Te,len(sta_ge), dt)
sta_th_filtered_ge = tf.whiten_filter(th_cov_ge,sta_ge)

lag_gi,sta_gi = tf.calculate_sta(dt, gi, spikes, window)
cov_gi = tf.estimate_cov_matrix(gi, len(sta_gi))
sta_filtered_gi = tf.whiten_filter(cov_gi,sta_gi)
th_cov_gi = tf.ou_cov_matrix(gi_sigma, Ti,len(sta_gi), dt)
sta_th_filtered_gi = tf.whiten_filter(th_cov_gi,sta_gi)


plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(lag_ge*1e3, sta_ge,'b');plt.xlabel('ms')
plt.twinx()
plt.plot(lag_ge*1e3, sta_th_filtered_ge,'r');plt.xlabel('ms')
plt.plot(lag_ge*1e3, sta_filtered_ge,'k--');plt.xlabel('ms')
#plt.plot(lag_gi*1e3, sta_gi);plt.xlabel('ms')
plt.subplot(122)
plt.plot(lag_gi*1e3, sta_gi,'b');plt.xlabel('ms')
plt.twinx()
plt.plot(lag_gi*1e3, sta_th_filtered_gi,'r');plt.xlabel('ms')
plt.plot(lag_gi*1e3, sta_filtered_gi,'k--');plt.xlabel('ms')
```

```python
spk_idx = (np.array(spikes)/dt).astype(int)
getotal, gespike = tf._calc_projections(ge, sta_filtered_ge, spk_idx)
gitotal, gispike = tf._calc_projections(gi, sta_filtered_gi, spk_idx)
```

```python
def calc_cov(x, y):
    x_nomean = x - x.mean()
    y_nomean = y - y.mean()
    x_matrix = np.vstack([x_nomean, y_nomean])
    C = np.dot(x_matrix, x_matrix.T)
    return C/len(x_nomean)
    
```

```python
_, evecs = np.linalg.eig(calc_cov(gespike, gispike))
```

```python
def linear_combination(x, y, v):
    return x*v[0]+y*v[1]
```

```python
vec = evecs[:, 0]
ggtotal = linear_combination(getotal, gitotal, vec)
ggspike = linear_combination(gespike, gispike, vec)

bins = np.linspace(ggtotal.min(), ggtotal.max(),10)
n_total, _ = np.histogram(ggtotal, bins)
n_spike, _ = np.histogram(ggspike, bins)
plt.plot(n_spike*1./n_total)
plt.ylim(0, 0.02)
```

```python
def norm(x):
    return x/np.sqrt(np.sum(x**2))
```

```python
### finding the rotation vector

def find_slope(pge_spikes, pgi_spikes):
    pge_spk = pge_spikes - pge_spikes.mean()
    pgi_spk = pgi_spikes - pgi_spikes.mean()
    pg = np.vstack((pge_spk,pgi_spk))
    cov = np.dot(pg,pg.T)
    cov /= len(pg)
    evals, evecs = np.linalg.eig(cov)
    return evecs[:,0]    
    
ge_proj, ge_spike = tf._calc_projections(ge, sta_filtered_ge, spk_idx)
gi_proj, gi_spike = tf._calc_projections(gi, sta_filtered_gi, spk_idx)

rot_vec = find_slope(ge_spike, gi_spike)
```

```python
vec = rot_vec
from scipy.interpolate import interp2d
plt.contourf(ge_bins, gi_bins, f_ge_gi,np.linspace(0,0.015,20))
xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()
r = xmax - xmin
ymax= 0.35e-7

for yshift in np.linspace(ymin, 2*ymax, 20):
    plt.plot([xmin, xmin+vec[0]*r], [yshift, yshift + vec[1]*r], 'w')
plt.ylim(ymin, ymax)
```

```python
rot_matrix = np.array([vec, [-vec[0], vec[1]]])
```

```python
def rotate(x, y):
    return np.dot(rot_matrix.T, np.array([x,y]))

nbins = 20
spk_idx= (np.array(spikes)/dt).astype(int)
ge_proj, ge_spike = tf._calc_projections(ge, sta_filtered_ge, spk_idx)
gi_proj, gi_spike = tf._calc_projections(gi, sta_filtered_gi, spk_idx)

ge_proj, gi_proj = rotate(ge_proj, gi_proj)
ge_spike, gi_spike = rotate(ge_spike, gi_spike)

ge_bins2 = np.linspace(ge_proj.min(), ge_proj.max(), nbins)
gi_bins2 = np.linspace(gi_proj.min(), gi_proj.max(), nbins)
n_total, _, _ = np.histogram2d(ge_proj, gi_proj,  [ge_bins2, gi_bins2])
n_spike, _, _ = np.histogram2d(ge_spike, gi_spike, [ge_bins2, gi_bins2])
f_val = n_spike/n_total
f_val[n_total==0] = 0
ge_bins2, gi_bins2 = ge_bins2[:-1], gi_bins2[:-1]
```
