```python
def PSP(t, Tm, Ts, J):
    if Ts==0:
       return J*np.exp(-t/Tm) 
    if Tm!=Ts:
        return J*Tm/(Tm-Ts)*(np.exp(-t/Tm)-np.exp(-t/Ts))
    else:
        return J*t/Tm*np.exp(-t/Tm)

t = np.linspace(0,50)
plt.figure()
plt.title('Effect of synaptic filtering on PSP')
plt.plot(t, PSP(t, 10, 0, 0.01), label=r"$\tau_S$ = 0 ms")
plt.plot(t, PSP(t, 10, 5, 0.01), label=r"$\tau_S$ = 5 ms")
plt.plot(t, PSP(t, 10, 10, 0.01), label=r"$\tau_S$ = 10 ms")
plt.xlabel('time from synaptic event')
plt.ylabel('membrane potential')
plt.legend()
```

```python
# we prepare the numerical functions with some functions used in the algorithm

def Tmax(Tm, Ts): # delay of the max jump value after a synaptic event
    if Tm!=Ts: 
        Tmax = Tm*Ts/(Ts-Tm) * np.log(Ts/Tm)
    else:
        Tmax = Tm
    return Tmax

from scipy.optimize import newton


def find_threshold_crossing(DV, Tm, Ts, J):
    # to be executed only if there is an actual threshold crossing V(tmax)>Vth
    if Tm!=Ts:
        fun = lambda x: J*Tm/(Tm-Ts)*(np.exp(-x/Tm)-np.exp(-x/Ts))-DV
    else:
        fun = lambda x: J*x/Tm*np.exp(-x/Tm)-DV

    res = newton(fun, threshold/4.)#, method='SLSQP', bounds=bnds)
    return res
    
J = 0.01
threshold = J/3.
Tm = 10.
Ts1 = 5.

t = np.linspace(0,50,1e2)
plt.figure(figsize=(10,5))
plt.title('Constructing the algorithm for the Event-based simulations')
plt.plot(t, PSP(t, Tm, Ts1, J), 'r', label=r"$\tau_S$ = 5 ms")
plt.plot(t, PSP(t, Tm, Tm, J), 'b', label=r"$\tau_S$ = 10 ms")
tmax, vmax = Tmax(Tm, Ts1), PSP(Tmax(Tm, Ts1), Tm, Ts1, J)
plt.plot([0, tmax, tmax], [vmax, vmax, 0],'r--', lw=2, label=r"$v_{max},\,\, t_{max}$")
tmax, vmax = Tmax(Tm, Tm), PSP(Tmax(Tm, Tm), Tm, Tm, J)
plt.plot([0, tmax, tmax], [vmax, vmax, 0],'b--', lw=2)
plt.plot(t, threshold*np.ones(t.size),'k-.',lw=2, label='threshold')
tthre = find_threshold_crossing(threshold, Tm, Ts1, J)
plt.plot([0, tthre, tthre], [threshold, threshold, 0],'r-', lw=2, label=r"$v_{thre}, \,\, t_{thre}$")
tthre = find_threshold_crossing(threshold, Tm, Tm, J)
plt.plot([0, tthre, tthre], [threshold, threshold, 0],'b-', lw=2)
plt.xlabel('time from synaptic event')
plt.ylabel('membrane potential')
plt.legend()
```

```python
## Numerical simulations of Brunel and Sergi
# now that we have all the quantities, we can run the simulation


vvec = []

def Shotnoise_Sim(nu, J, Tm, Ts, v_thre, v_reset):
    if nu==0: print "NON ZERO FREQUENCY NEEDED!!"
    # they say nearly 1e8 events, we will take 1e7
    
    # we prepare the pre-synaptic spike train
    pre_spk_train = np.random.exponential(1./nu, 1e4)
    pre_spk_train = np.cumsum(pre_spk_train)
    
    counter_spk = 0 # spike counter
    v = 0 # membrane potential at last pre-syn event
    t_last = pre_spk_train[0] # time at last pre-syn event
    for pre_spk in pre_spk_train[1:]:
        # we start by testing the effect of the last spike
        # did this last event could trigger a spike ??
        if PSP(Tmax(Tm, Ts), Tm, Ts, J)+v>v_thre:
            # if it can, when  would it trigger it ?
            tspk = find_threshold_crossing(v_thre-v, Tm, Ts, J)
        else:    
            tspk = np.inf # this means no post-syn spike !
        # so now, we test if there is no pre-syn spike, that happens before the post-syn spike
        if tspk<pre_spk-t_last: # then SPIKE
            counter_spk +=1
            v = v_reset # we reset the potential
        else:
            v = PSP(pre_spk, Tm, Ts, J) + (1-v) * np.exp(-pre_spk/Tm) # we update the potential
        # then, whatever happens, we update the time to the next spike, as nothing can happen without any event
        # (no leak because El = v_reset)
        t_last += pre_spk
        vvec.append(v)
    return counter_spk/pre_spk_train[-1]
        
       
Shotnoise_Sim(90., 0.01, 0.010, 0.005, 1., 0.)   
```

```python
from scipy import special, integrate

def gauss_integral(x, mu, sigma):
    """integral of normal distribution from x to inf"""
    return special.erfc((x-mu)/np.sqrt(2)/sigma)/2.

def gaussian_1d(x, mu, sigma):
    return 1./(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2.*sigma**2))

def erf_integral(x, mu, sigma):
    """integral from -inf to x of gauss_integral"""
    x_ = (x-mu)/sigma/np.sqrt(2)
    return (x_*(1+special.erf(x_))+np.exp(-x_**2)/np.sqrt(np.pi))*sigma*np.sqrt(2)

def ou_rate_markov_anal(mu, sigma, tau, threshold):
    sigma_markov = np.sqrt(sigma**2/(1./tau_m*(1./tau_m+1./tau)))
    mu_markov = (mu*tau_m)
    correction = np.sqrt((tau+tau_m)/tau_m)
    
    stationary = gaussian_1d(threshold, mu_markov, sigma_markov)
    markov_integral = erf_integral(0, (threshold), sigma)/2.
    prob = markov_integral*stationary/correction
    return prob

def ou_rate_markov_anal2(mu, sigma, tau, threshold):
    sigma_markov = np.sqrt(sigma**2/(1./tau_m*(1./tau_m+1./tau)))
    mu_markov = (mu*tau_m)
    
    stationary = gaussian_1d(threshold, mu_markov, sigma_markov)
    markov_integral = erf_integral(0, (threshold-mu_markov)/np.sqrt(tau), sigma_markov/tau)/2.
    prob = markov_integral*stationary
    return prob
```

## Brunel Functions

```python
import scipy.integrate as sp_int
import scipy.special as sp_spec
def Psi(Array):
    return np.exp(Array*Array)*(sp_spec.erf(Array)+1)

## First passage time of an Ornstein Uhlenbeck process
# (membrane equation with white noise input)
## without correction because of the colored noise we would get
def BrunelFunc(mVm, sVm, Tm, Vthre, Vreset):
    x1=(Vreset-mVm)/(sVm)
    x2=(Vthre-mVm)/(sVm)
    integral = sp_int.quad(Psi,x1,x2)
    if integral[0]<float('inf') and integral[0]>0:
        T=Tm*np.sqrt(np.pi)*integral[0]
        freq=1.0/T
    else:
        freq=1e-9
    return freq
## now with correction for correlated noise input 
# (membrane equation with colored noise input, correlation time : Tau )
def BrunelFunc_colored(mVm, sVm, Tm, Vthre, Vreset, Tau):
    # taken from Brunel and Sergi 1998 :
    return BrunelFunc(mVm, sVm, Tm, Vthre + 1.03*np.sqrt(Tau/Tm)*sVm, Vreset)

sVm = np.sqrt(sigmas**2/(1./tau_m*(1./tau_m+1./tau)))
brunel0 = np.array([(BrunelFunc(mVm, _, tau_m, threshold, reset))
                      for _ in sVm]).T
brunel1 = np.array([(BrunelFunc_colored(mVm, _, tau_m, threshold, reset, tau))
                      for _ in sVm]).T
```
