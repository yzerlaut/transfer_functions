```python
import numpy as np
import matplotlib.pylab as plt
from scipy.optimize  import leastsq

from my_env import * # to be removed, nothing important, just some esthetic things for the graphs

## PHYSICAL UNITS ###

# parameters
Gl= 1./60*1e-6 ; Cm= 250*1e-12 ; El = -70*1e-3 
Qe = 7.1e-9 ; Te = .2e-3; Ee = 0e-3 
Qi = 3.7e-9 ; Ti = 2e-3; Ei = -75e-3
Vthre=-50e-3; Vreset = -60e-3 ; refrac = 0 ; 

import process_generators as pg
import analytical_formula as af
from scipy.special import erfc
dt = 1e-5
tmax = .1
```

## Study equivalent to Kuhn et al. 2004 but with exponential synapses (to use OU stochastic processes)

```python
## stimulation, see Kuhn et al. for details...
v_clamp = -55e-3 # we ajust the stimulation so that it is clamps the mean membrane potential to this value

log_fe0 = np.log(-(El-v_clamp)/(Ee-v_clamp) * Gl/Qe / Te +1e-9)/np.log(10) # fe for fi=0
fe = np.logspace(log_fe0,5,3)

fi = -(Ee-v_clamp)/(Ei-v_clamp) * Qe/Qi * Te/Ti *fe - (El-v_clamp)/(Ei-v_clamp) * Gl/Qi / Ti # COMODULATION of fi and fe

# storing the relevant variables
vm_firing_traces = []
vm_free_pot_traces = []
cv_isi = []
rate = []
plt.figure(figsize=(14,4))

for ii in range(fe.size):
    
    # simulation with spikes
    gexc, exc_list, spk_exc_list = pg.conductance_shotnoise(pg.poisson_spike_generator(fe[ii]), 'exp', Ee, T=Te, Q=Qe)
    ginh, inh_list, spk_inh_list = pg.conductance_shotnoise(pg.poisson_spike_generator(fi[ii]), 'exp', Ei, T=Ti, Q=Qi)
    t, v_m, spikes = pg.leaky_iaf(tmax, dt, gexc, ginh,
        Cm=Cm, Gl=Gl, El=El, Vthre=Vthre, Vreset=Vreset, t_refrac=2e-3, max_spikes=10)
    # plt.plot(t,v_m) ## just to check the plotting
    rate.append(len(spikes)/t.max()) # rate 
    vm_firing_traces.append(v_m) # traces with spikes
    
    if len(spikes)>2:
        isi = np.diff(spikes)
        out = isi.std()/isi.mean()
    else:
        out=0
    cv_isi.append(out)

    # simulation for the free membrane potential, 500ms are largely sufficient !!
    gexc, exc_list, spk_exc_list = pg.conductance_shotnoise(pg.poisson_spike_generator(fe[ii]), 'exp', Ee, T=Te, Q=Qe)
    ginh, inh_list, spk_inh_list = pg.conductance_shotnoise(pg.poisson_spike_generator(fi[ii]), 'exp', Ei, T=Ti, Q=Qi)
    t, v_m, spikes = pg.leaky_iaf(.5, dt, gexc, ginh, Cm=Cm, Gl=Gl, El=El, Vthre=1) # i.e. no spiking (high threshold)

    vm_free_pot_traces.append(v_m)
```

### Analysis

```python
# mean and variances
mean_vm = []
std_vm = []

## autocorrelation functions (to mesure the effective time correlation constant)
time_lag = int(30e-3/dt) # integer to evaluate the time lag
lag = -np.arange(time_lag+1)[::-1]*dt*1e3 # ms
autocorr_time = []

# fitting of the autocorrelation time constant 
expfunc = lambda p, x: p[1]*np.exp(-abs(x)/p[0])+p[2] # Exponential function
errfunc = lambda p, x, y: expfunc(p, x) - y # Distance to the exponential function
    
plt.figure(figsize=(7,4))
plt.title('autocorrelation function')
plt.xlabel('time lag')
plt.plot([-30,0],[0,0],'k--',lw=3,alpha=.5)
for ii in range(fe.size):
    yy = vm_free_pot_traces[ii][int(20e-3/dt):] # we discard the first 20 ms, exc is faster than inh -> unbalanced transients !!
    # mean and variance
    mean_vm.append(yy.mean())
    std_vm.append(yy.std())
    # autocorrelation
    yy = (yy - yy.mean())/yy.std()
    autocorr = np.convolve(yy[:],yy[time_lag:][::-1],mode='valid')/yy[time_lag:].size
    # fitting 
    p0 = [10., 1., 0.]
    p,success = leastsq(errfunc, p0[:], args=(lag, autocorr))
    autocorr_time.append(p[0]) # ms !! we store the time decay constant
    ## plotting
    red = (float(fe[ii])-fe.min())/(fe.max()-fe.min())
    plt.plot(lag, autocorr,'-',lw=2,color=(red,0,1-red))
    plt.plot(lag, expfunc(p,lag),'-',lw=4,alpha=.5,color=(red,0,1-red))
```

```python
## theoretical calculations
v_clamp = -55e-3
log_fe0 = np.log(-(El-v_clamp)/(Ee-v_clamp) * Gl/Qe / Te +1)/np.log(10) # fe for fi=0
fe_th = np.logspace(log_fe0,5,1e2)

fi_th = -(Ee-v_clamp)/(Ei-v_clamp) * Qe/Qi * Te/Ti *fe_th - (El-v_clamp)/(Ei-v_clamp) * Gl/Qi / Ti

mean_vm_rudolph = []
param_rudolph = { 'Cm' : Cm, 'Gl' : Gl, 'El' : El, 'Vthre' : Vthre,
                  'Te' : Te, 'Ee' : Ee, 'Ti' : Ti, 'Ei' : Ei }


time_cst = [] # first order approximation of the time constant
std_time_cst = [] # theoretical variance around it
    
std_vm_kuhn = [] # basic kuhn approach (kuhn et al. 2004)
std_vm_rudolph = []
rate_kr = [] # kuhn firing rate estimate + rudolph membrane computation
rate_kuhn = []

for ii in range(fe_th.size):
    
    # quantities for analytical calculations
    mu_ge, sigma_ge = af.campbell_theorem_single_exp(fe_th[ii], Qe, Te)
    mu_gi, sigma_gi = af.campbell_theorem_single_exp(fi_th[ii], Qi, Ti)
    
    Tm = Cm/(mu_ge+mu_gi+Gl)
    time_cst.append(Tm*1e3) # ms
    std_time_cst.append( 1e3*Tm*np.sqrt(sigma_ge**2+sigma_gi**2)/(mu_ge+mu_gi+Gl)) # ms
    
    ## PSP square gives the membrane potential variance (under Kuhn approx, fixed Tm and driving force )
    EPSP2 = min( Te * ( (Ee-v_clamp)*Qe/Cm*Tm )**2, Te/2 * ( ( (Ee-v_clamp)*Qe*Tm*Te )/(Cm*(Te-Tm) ) )**2 )# square integral of the PSP
    IPSP2 = min( Ti * ( (Ei-v_clamp)*Qi/Cm*Tm )**2, Ti/2 * ( ( (Ei-v_clamp)*Qi*Tm*Ti )/(Cm*(Ti-Tm) ) )**2 ) 
    # this take care of the Tm=Ti pb !!!!
    
    sigma_v_kuhn  = np.sqrt( fe_th[ii]*EPSP2 + fi_th[ii]*IPSP2 )
    
    std_vm_kuhn.append(sigma_v_kuhn)
    rate_kuhn.append(erfc((Vthre-v_clamp)/np.sqrt(2)/sigma_v_kuhn)/2/Tm)
    
    # rudolph quantities
    param_rudolph['mu_ge'], param_rudolph['s_ge'] = mu_ge, sigma_ge
    param_rudolph['mu_gi'], param_rudolph['s_gi'] = mu_gi, sigma_gi
    rate_kr.append(af.proba_rudolph(param_rudolph))
    #print af.proba_rudolph(param_rudolph)
```

```python
plt.subplot(111)
plt.semilogx(fe_th,1e3*np.array(std_vm_kuhn),'k--',lw=4,alpha=.4, label='Kuhn approx')
plt.ylabel(r"free $V_m$ std")
plt.xlabel(r" $\nu_e$ (Hz)")
plt.legend(loc='upper left', frameon=False, prop={'size':16})
```
