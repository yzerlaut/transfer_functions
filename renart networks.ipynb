```python
import numpy as np

binsz = 0.1e-3
fr_in = 20.
r_in = 0.025
t_max = 10.
Nexc, Ninh = 250, 220
p = 0.2
```

```python
mother_train = (np.random.rand(int(t_max/binsz))>fr_in*binsz)*1
```

```python
from scipy import stats
def generate_poisson(fr, tmax):
    t = np.cumsum(stats.expon.rvs(scale=1./fr, size=int(tmax*fr*1.2)))
    t = t[:t.searchsorted(tmax)]
    return t

def generate_correlated(mother, r):
    rn = np.random.rand(len(mother))
    return mother[rn<r]

def generate_current(spt, binsz, tmax, tau):
    npts = tmax/binsz
    current = np.zeros(npts, dtype=float)
    kernelsz = int(tau/binsz)*10
    kernel = np.exp(-np.arange(kernelsz)*binsz*1./tau)
    idx = (spt/binsz).astype(int)
    #idx = idx[idx<(npts-kernelsz)]
    for i in idx:
        f, l = min(max(0,i), npts-1), min(i+kernelsz, npts-1)
        current[f:l]+=kernel[f-i:l-i]
    return current
    
def jitter_spikes(spt, tau):
    sign = 1-2*np.random.rand(len(spt))
    delta_t = stats.expon.rvs(scale=tau, size=len(spt))*sign
    return spt+delta_t
    
    

mother = generate_poisson(fr_in/r_in, t_max)
corr_train = generate_correlated(mother, r_in)
corr_train = jitter_spikes(corr_train, 5.e-3)
current = generate_current(corr_train, binsz, t_max, 10.e-3)
```

```python
def current_generator(N, mother, r_in, tau_syn):
    jitter = 5.e-3
    for _ in xrange(N):
        corr_train = generate_correlated(mother, r_in)
        corr_train = jitter_spikes(corr_train, jitter)
        current = generate_current(corr_train, binsz, t_max, tau_syn)
        yield current[np.int(tau_syn/binsz)*10:]
        

mother = generate_poisson(fr_in*1./r_in, t_max)
def two_cell_syn(N, p, mother, r_in, tau_syn):
    i_exc_1 = sum(current_generator(int(N*(1-p)), mother, r_in, 10.e-3)) 
    i_exc_2 = sum(current_generator(int(N*(1-p)), mother, r_in, 10.e-3)) 
    i_exc_c = sum(current_generator(N-int(N*(1-p)), mother, r_in, 10.e-3)) 
    return i_exc_1 + i_exc_c,  i_exc_2 + i_exc_c

i_exc1, i_exc2 = two_cell_syn(Nexc, p, mother, r_in, 10.e-3)
i_inh1, i_inh2 = two_cell_syn(Ninh, p, mother, r_in, 10.e-3)
```

```python
def plot_currents(i_exc, i_inh, **kwargs):
    plt.plot(i_exc, color='r', **kwargs)
    plt.plot(-i_inh, color='b', **kwargs)
    plt.plot(i_exc-i_inh, color='k')
plot_currents(i_exc1, i_inh1)
plot_currents(i_exc2, i_inh2, ls='--')
plt.xlim([0, 2000])
```

```python
def xcorr(x, y, maxlag):
    corr = np.convolve(x[:]-x.mean(), y[-maxlag:maxlag:-1]-y.mean(), 'valid')
    return corr/(len(x-2*maxlag))/np.std(x)/np.std(y)

plt.plot(xcorr(i_exc1-i_inh1, i_exc2-i_inh2, 100))
plt.plot(xcorr((i_exc1+i_exc2)/2., i_exc1-i_inh1, 100))
```
