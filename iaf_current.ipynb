```python
import numpy as np
import matplotlib.pylab as plt

## PHYSICAL UNITS ###

# parameters
Gl= 1e-8 ; Cm= 0.2*1e-9; El = -65*1e-3 
Ne = 200 ; Qe = 6*1e-9 ; Te = 5*1e-3; Ee = 0*1e-3 ; fe = 4
Ni = 50 ; Qi = 64*1e-9 ; Ti = 10*1e-3; Ei = -80*1e-3 ; fi= 10
Vthre=-50*1e-3; refrac = 5*1e-3 ; 

import tf_filters as tf
```

```python
#### TESTING THE EFFECT OF GAUSSIAN WHITE NOISE ####

dt = 1e-4
tstop = 1000

# current stimulation values
I0 = 100*1e-12 # Ampere
sI = 400*1e-12 # A

# we try a simulation with ~ntwk sim values
i = tf.white_gaussian(tstop, dt, I0, sI)

t, v, spikes = tf.launch_sim_current(dt, i,spiking_mech=True,max_spikes=300)
print " spike number : ",len(spikes)
print " firing frequency : ",1.*len(spikes)/tstop

plt.figure(figsize=(15,4))
plt.plot(t[:1e4]*1e3,v[:1e4]*1e3)
```

```python
window = (-50*1e-3,0)
lag_i,sta_i = tf.calculate_sta(dt, i-i.mean(), spikes, window)

plt.figure()
plt.plot(lag_i*1e3, sta_i,label='STA filter')
plt.xlabel('ms')
# we compare it with the membrane green function 
lag_th = np.arange(0,1000)*dt #ready in ms
green_func = np.exp(-lag_th*Gl/Cm)*(sta_i.max()-sta_i.min())
plt.plot(lag_th[::-1]*1e3-100, green_func,label='Green function')
plt.legend(loc='upper left')
plt.xlim([-50,0])
```

```python
## Non linearity 1D

i_bins, i_spike, i_total = tf.estimate_nonlinearity_1d(i, sta_i, spikes, dt)
plt.subplot(221); plt.plot(i_bins, i_spike)
plt.axvline(color='k')
plt.title('spike-triggered ensemble')
plt.ylabel('excitation')
plt.subplot(222); plt.plot(i_bins, i_spike*1./i_total, '.')
plt.ylim((0,1e-4))
plt.title('nonlinearity')
```

```python
## Non linearity 1D with respect to the green function

i_bins, i_spike, i_total = tf.estimate_nonlinearity_1d(i, green_func[::-1], spikes, dt)
plt.subplot(221); plt.plot(i_bins, i_spike)
plt.axvline(color='k')
plt.title('spike-triggered ensemble')
plt.ylabel('excitation')
plt.subplot(222); plt.plot(i_bins, i_spike*1./i_total, '.')
plt.ylim((0,1e-4))
plt.title('nonlinearity')
```
